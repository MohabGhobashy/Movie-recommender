```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.impute import KNNImputer

```

```python
import pandas as pd
movies_df = pd.read_csv("movies.csv")
ratings_df = pd.read_csv("ratings.csv")
movies_df = movies_df.drop([ 'genres'], axis=1)
ratings_df = ratings_df.drop(['timestamp'], axis = 1)
```

```python
merged_df = pd.merge(movies_df, ratings_df, on='movieId')
merged_df.head()
```

```python
sorted_movie_titles = movies_df['title'].tolist()

# The above code retrieves the list of movie titles from the 'title' column of the movies_df DataFrame
# and converts it into a Python list.


merged_df = merged_df.pivot_table(index='userId', columns='title', values='rating', sort=True)

# The pivot_table function is used to reshape the merged_df DataFrame.
# It sets the 'userId' column as the index, 'title' column as the columns,
# and 'rating' column as the values of the new DataFrame.


merged_df = merged_df.reindex(columns=sorted_movie_titles)

# The reindex function is used to reorder the columns of the merged_df DataFrame based on the sorted_movie_titles list.
# It ensures that the columns (movie titles) are in the desired order.


merged_df.head()

# The above code displays the first few rows of the merged_df DataFrame.


```

```python
# Filtering the users to the first 200 user, and 200 movies
selected_users = merged_df.iloc[:200]
selected_users = selected_users.iloc[:, :200]
selected_users
```

```python

selected_users=selected_users.fillna(0)
selected_users
```

```python

# Calculate cosine similarity
cosine_sim = cosine_similarity(selected_users.T)
cosine_sim_df = pd.DataFrame(cosine_sim, columns=selected_users.columns, index=selected_users.columns)
```

```python
cosine_sim_df.head()
```

```python
def get_similar_movies(movie_title, num_movies=10):
    # Get the cosine similarity row for the given movie title
    movie_row = cosine_sim_df.loc[movie_title]
    
    # Sort the row in descending order
    sorted_row = movie_row.sort_values(ascending=False)
    
    # Exclude the movie itself
    sorted_row = sorted_row[1:]
    
    # Get the top N movies
    top_movies = sorted_row.head(num_movies)
    
    # Retrieve the movie titles and create a DataFrame
    similar_movies = pd.DataFrame({'Title': top_movies.index})
    
    # Return the DataFrame
    return similar_movies




```

```python
toy_story_similar_movies = get_similar_movies('Toy Story (1995)', num_movies=10)

# Display the similar movies
toy_story_similar_movies
```

```python
user_200 = selected_users.loc[200]

user_200_movies = []
for column_name, value in user_200.iteritems():
    if value > 0:
        user_200_movies.append((column_name, value))
len(user_200_movies)

# The above code selects the movies that user_200 has rated positively (rating > 0)
# and stores them in the user_200_movies list.


def similarity_function(movie, rating, watched_movies):
  
  # Less weight given for movies with a rating lower than 3 
  similarity_score = cosine_sim_df[movie] * (rating - 3)
  similarity_score = similarity_score.sort_values(ascending=False)

  similarity_score = similarity_score[~similarity_score.index.isin(watched_movies)]  # Exclude watched movies

  return similarity_score

# The similarity_function takes a movie, its rating, and a list of watched movies as input.
# It calculates the similarity score between the given movie and all other movies using cosine similarity.
# The similarity score is adjusted based on the rating (less weight for ratings lower than 3).
# The function returns the similarity scores for the unwatched movies.


similar_movies = pd.DataFrame()

# Creating a list with the already watched movies by the user
watched_movies = [movie for movie, _ in user_200_movies]

# The watched_movies list contains the names of movies that user_200 has already watched.


for movie, rating in user_200_movies:
    similar_movies = similar_movies.append(similarity_function(movie, rating, watched_movies)).reset_index(drop=True)

# The above code iterates over each movie in user_200_movies and calls the similarity_function for each movie.
# It appends the similarity scores for unwatched movies to the similar_movies DataFrame.


similar_movies.head()

top_3_indices = similar_movies.sum().sort_values(ascending=False).head(3).index

# The above code calculates the total similarity score for each movie by summing up the similarity scores across all users.
# It then selects the top 3 movies with the highest similarity scores.


print("Recommended movies:")
for movie in top_3_indices:
  print(movie)

# The above code prints the top 3 recommended movies based on the highest total similarity scores.

```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

movies=pd.read_csv('movies.csv')
rating=pd.read_csv("ratings.csv")


title = movies["title"].tolist()
retyped_titles = [re.sub(r"[^a-zA-Z0-9] ", "", str(t).lower()) for t in title]
movies["retyped_titles"] = retyped_titles
movies

vectorizer = TfidfVectorizer(ngram_range=(1, 2))
vectorized_titled = vectorizer.fit_transform(movies["retyped_titles"])

def text_preprocessing(new_input):
    new_input = re.sub("[^a-zA-Z0-9 ]", "", new_input).lower()
    new_input = vectorizer.transform([new_input])
    return new_input


def recommendation_system(new_input):
    vectorized_input = text_preprocessing(new_input)

    # Compute cosine similarity between vectorized titles and vectorized input
    similarity = cosine_similarity(vectorized_titled, vectorized_input).flatten()

    # Get the indices of the top 10 similar movies
    indices_top_10 = np.argpartition(similarity, -10)[-10:]

    # Get the movie ID of the most similar movie
    movie_id = movies.iloc[indices_top_10[::-1]]['movieId'].values[0]

    # Get users who rated the most similar movie highly (rating > 4)
    users_with_same_interest_on_film_df = rating[(rating["movieId"] == movie_id) & (rating['rating'] > 4)]
    users_with_same_interest_on_film = users_with_same_interest_on_film_df["userId"].unique()

    # Get other movies rated highly by users with similar interests
    users_other_recommendation_df = rating[(rating['userId'].isin(users_with_same_interest_on_film)) & (rating['rating'] > 4)]
    users_other_recommendation_df_filmsId = users_other_recommendation_df['movieId']

    # Calculate the percentage of interest for each movie
    percent_of_interest = users_other_recommendation_df_filmsId.value_counts() / len(users_with_same_interest_on_film)

    # Filter movies with interest percentage greater than 0.2
    filtered_movies = percent_of_interest[percent_of_interest > 0.2]

    # Get all users who rated the filtered movies highly
    all_people = rating[(rating["movieId"].isin(filtered_movies.index)) & (rating['rating'] > 4)]

    # Calculate the percentage of users interested in each movie
    all_people_percent = all_people["movieId"].value_counts() / len(all_people["userId"].unique())

    # Combine the two percentage series into a single dataframe
    rec_percentages = pd.concat([filtered_movies, all_people_percent], axis=1)
    rec_percentages.columns = ["similar_interest", "all"]

    # Calculate the score for each movie based on the ratio of similar interest to total interest
    rec_percentages["score"] = rec_percentages["similar_interest"] / rec_percentages["all"]

    # Sort the movies based on the score in descending order
    rec_percentages = rec_percentages.sort_values("score", ascending=False)

    # Get the top 10 recommended movies and merge with movie information
    final_df = rec_percentages.head(10).merge(movies, left_index=True, right_on="movieId")

    return final_df, movie_id


movie1 = 'toy story'
movie2 = 'Waiting to Exhale'
final_df, movie_id = recommendation_system(movie1)
final_df2, movie_id2 = recommendation_system(movie2)

print("Recommendations for movie:", movie1)
display(final_df)

print("\nRecommendations for movie:", movie2)
display(final_df2)
```

```python
def recommend_films_to_user(user_id, top_k=3):
    # Get ratings of the specified user
    user_ratings = rating[rating['userId'] == user_id]

    # Get films rated highly (rating > 4) by the user
    films_rated = user_ratings[user_ratings['rating'] > 4]['movieId']

    # Initialize list to store recommended films
    recommended_films = []

    # Iterate over films rated by the user
    for film_id in films_rated:
        # Get the title of the film
        film_title = movies[movies['movieId'] == film_id]['retyped_titles'].values[0]

        # Get recommendations for the film using the recommendation_system function
        recommendations, _ = recommendation_system(film_title)

        # Add the movie IDs of the recommended films to the list
        recommended_films.extend(recommendations['movieId'])

    # Calculate recommendation scores for the films
    rec_scores = pd.Series(recommended_films).value_counts()

    # Get the top-k films with the highest recommendation scores
    top_films = rec_scores.nlargest(top_k).index.tolist()

    # Return the movie information for the top recommended films
    return movies[movies['movieId'].isin(top_films)]



recommended_films_user_200 = recommend_films_to_user(200, top_k=3)
print("Recommended films for userID 200:")
display(recommended_films_user_200)
```
